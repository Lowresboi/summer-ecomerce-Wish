# DATA CLEANING

```python
# Checking the nulls by columns
data_products.isnull().sum()
```

```python
# Define the categorical columns to process
categorial_to_process = ['has_urgency_banner', 'urgency_text']

# Fill the columns with a loop
for column in categorial_to_process:
  data_products[column] = data_products[column].fillna('F')
```

```python
# Define the numerical columns to process
numerical_to_process = ['rating_five_count', 'rating_four_count', 'rating_three_count', 'rating_two_count', 'rating_one_count']

# Fill the column with a loop
for column in numerical_to_process:
    data_products[column] = data_products[column].fillna(data_products[column].mean())
```

```python
# Define the last numerical columns to process
last_columns_to_process = ['product_color', 'product_variation_size_id', 'origin_country', 'merchant_name', 'merchant_info_subtitle']

# Fill the last columns with a loop
for column in last_columns_to_process:
    data_products[column].fillna(data_products[column].mode()[0], inplace=True)
```

```python
# Dropping the last columns with urls
columns_to_drop = ['merchant_profile_picture', 'product_url', 'product_picture']

data_products.drop(columns=columns_to_drop, inplace=True)
```

```python
# Identifing the duplicated data and droping
duplicated_data = data_products.duplicated()
sum(duplicated_data)
```

```python
# Droping the duplicates
data_p1 = data_products.drop_duplicates()
```

```python
# Now the data its clean from duplicates
duplicated_data = data_p1.duplicated()
sum(duplicated_data)
```
